```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('timing.csv', sep=';', header=None, index_col=False)

plt.figure(figsize=(10,5))
plt.plot(df[0],df[1],'k')
plt.title('Errorbar of time spent in average based on the number of workers')
plt.xlabel('Number of workers')
plt.xticks(df[0])
plt.ylabel('Time spent (avg)')
plt.errorbar(df[0],df[1],yerr=df[2], capsize = 2, fmt = 'ro', label = 'standard deviation')

plt.savefig('Errorbar.JPG')

plt.show()
```
